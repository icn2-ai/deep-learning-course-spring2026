## Concepts


### Variables

A variable is something you define in your code with a name:

```python
x = 1
float_number = 3.14
my_name = "Ryland Grace"
```

First comes the name of the variable, then what you assign to it. Python is pretty flexible when defining variables of different type: integer, float, strings, etc.

### Lists

You can define lists of variables too:

```python
integers = [1, 2, 3, 4]
floats = [1.2, 2.3, 3.4]
numbers = [1, 1.2, 3.4, 5]
strings_and_numbers = ["Hello", 2, 4, "Bye"]
```

Again, Python is very flexible with these. We would recommend that you stick to the same type of variable of all elements in a list. You can access their elements with indexing (first element will always have index 0, not 1):

```python
print(strings_and_numbers[0])
```

### Conditionals

We can use conditions in our code to run certain commands depending on whether if the input meets certain criteria.

```python
distance = 5000

if distance > 2500:
  print("We are safe!")
else:
  print("We are going to crash!")
```

```python
animal = "cat"

if animal == "dog":
  print("I love dogs!")
elif animal == "cat":
  print("I am alergic to cats.")
else:
  print("I don't know this animal.")
```

### Loops

We use loops when there is an operation that we have to repeat.

```python
for i in range(10):
  print(i)
```

```python
names = ["Ryland Grace from Earth", "Rocky from Erid"]
for name in names:
  print(hello(name))
```

### Function

A function is a piece of code that receives some arguments and returns a value:

```python
# You can define the types of the arguments and the return value
def simple_sum(a: int, b: int) -> int:
  return a + b

print(simple_sum(2, 3))
```

```python
def hello(name: str) -> str:
  return "Hello, " + name + "!"

print(hello("Ryland"))
```

We use functions when there is a piece of code that we want to use several times, so that we do not have to repeat the code used.

### Class

Python is an object-oriented language, which allows us to structure the code using classes and objects for better organization and reusability.

For example, a class could be Animal and an object of that class would be dog. In a class, you have the `__init__` method, that allows you to initialize an object given a set of parameters. With those parameters, you initialize the internal *attributes* of the class, e.g., `self.something`. You can also define internal methods for the class, which are functions you can call if you have declared an object of that class.

```python
class Animal:
  def __init__(self, name: str, legs: int, sound: str):
    # Initialization of internal parameters
    self.name = name
    self.legs = legs
    self.sound = sound

  def make_sound(self):
    return self.sound + "!"
```

To access the attributes or methods, you can just call them after using a dot on the declared object.

```python
dog = Animal("dog", 4, "woof")
print(dog.name)
print(dog.legs)
print(dog.make_sound())
```
